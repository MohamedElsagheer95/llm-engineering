# **AI Multi-Language SRT Subtitle Translator**

This Python app translates .srt subtitle files from English to any language you want using local LLMs (Ollama)and  gemini cloud model, while preserving timestamps and subtitle structure.




>We parse subtitle files using `pysrt`, extract each subtitle block, and send only the text to an LLM for translation. The translated text is then reinserted into the original structure, ensuring perfect formatting and timing.

```python
#creating Ollama client

OLLAMA_BASE_URL = "http://localhost:11434/v1"

ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
print("your ollama client is ready to use")
```

```python

def translate_text(text,language):
    """Translates a single subtitle string  using an LLM."""
    if not text.strip():
        return ""
    
    response = ollama.chat.completions.create(
        model=model_llama,
        messages=[
            {"role": "system", "content": f"Translate the following subtitle text from English to {language} . Return only the translated text, nothing else."},
            {"role": "user", "content": text}
        ],
        temperature=0.3
    )
    return response.choices[0].message.content
```

```python

def translate_srt(input_path, output_path,language):
    # Load the original English file
    subs = pysrt.open(input_path)
    
    print(f"Starting translation of {len(subs)} blocks...\n")
    
    for i, sub in enumerate (subs):
        # Translate the text of the specific block
        print(f"Translating {i + 1}/{len(subs)} blocks...")
        print("original text: "+sub.text)
        translated_text = translate_text(sub.text,language)
        sub.text = translated_text
        print("translated text: "+translated_text+"\n")
        

    # Save the new file
    subs.save(output_path, encoding='utf-8')
    print(f"\n Success! {language} SRT saved to: {output_path}")
```
